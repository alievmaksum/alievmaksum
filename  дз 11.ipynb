```python
df.loc[7: 633, ['age','case']] 
# df переменная, содержащая DataFrame (таблицу данных).
#.loc[...]: Это метод доступа к данным в DataFrame по меткам (индексам) строк и столбцов. В квадратных скобках указываются критерии выбора.
# цыфры это спез строки DataFrame и имеют конечный индекс от7 включительно до 633 включительно 
# ['age','case']:это название столбцов которые нужно выбрать в данном случие это возраст и изучаемое явление 
```

```python
X = df.loc[:,['age']]
X
# переменная X стоновить age весь столбец 
```

```python
Y = df.loc[:,['case']]
X = df.loc[:,['age']]
# переменная Y становиться case весь столбец 
# переменная X становиться age  весь столбец 
```

```python
plt.figure(figsize=(7,3))
plt.plot(X,Y)
plt.show()
# plt.figure(figsize=(7,3)): Эта строка создает объект фигуры (окно графика) с заданными размерами. 
#figsize=(7,3) устанавливает ширину графика равной 7 дюймам и высоту — 3 дюймам.
#plt.plot(X,Y) это основная команда которая принимает из столбца X и Y 
#ункция plot строит график, отмечая точки с координатами (X[i], Y[i]) и соединяя их линией.
# plt.show(): Эта строка отображает построенный график. Без этой команды график будет построен, но не будет показан на экране.
```

```python
from sklearn.linear_model import LinearRegression
# from sklearn.linear_model: 
#Эта часть указывает путь к нужному классу. sklearn — это популярная библиотека машинного обучения в Python.
# linear_model — это модуль внутри sklearn, содержащий классы для построения различных моделей линейной регрессии.
# import LinearRegression: 
#Эта часть импортирует конкретный класс LinearRegression. Этот класс используется для построения модели линейной регрессии.
```

```python
model = LinearRegression()
model.fit(X,Y)
#model = LinearRegression(): Создается объект модели линейной регрессии.
#inearRegression() — это конструктор класса LinearRegression, который инициализирует модель с параметрами по умолчанию. 
#Этот объект хранится в переменной model.
#model.fit(X, Y): Эта строка обучает (настраивает) модель линейной регрессии на данных.
  # model.fit() — это метод объекта model, который выполняет обучение.
 # X — это матрица признаков (независимые переменные), предположительно, DataFrame или NumPy массив.
 # Y — это вектор целевых переменных (зависимая переменная), также, предположительно, Series или NumPy массив.
# В результате выполнения этого кода в переменной model хранится обученная модель линейной регрессии. 
# Она может быть использована для предсказания значений зависимой переменной Y на основе новых значений независимых переменныхX.

```

```python
model.coef_ 
# model.coef_ возвращает массив коэффициентов регрессии. 
#Эти коэффициенты показывают влияние каждой независимой переменной на зависимую переменную.
 #Если у вас одна независимая переменная, model.coef_ будет скалярным значением (одним числом), 
#представляющим наклон прямой линии регрессии.

#Если у вас несколько независимых переменных, model.coef_ будет массивом (вектором), 
#где каждый элемент массива соответствует коэффициенту для соответствующей независимой переменной.
```

```python
model.intercept_
#model.intercept_ представляет собой свободный член (или точку пересечения с осью Y) уравнения регрессии.
#Это значение показывает прогнозируемое значение зависимой переменной, когда все независимые переменные равны нулю. 
#Графически, это точка, где прямая (или гиперплоскость в многомерном случае) регрессии пересекает ось Y.
```

```python
y_predict = model.predict(X)
y_predict
#y_predict — это массив (или Series pandas, в зависимости от типа X) предсказанных значений зависимой переменной.
#Команда y_predict = model.predict(X) использует обученную модель линейной регрессии (model), 
#чтобы сделать предсказания на основе значений независимых переменных, хранящихся в X. 
#Таким образом, y_predict содержит значения, которые модель предсказывает для зависимой переменной на основе данных в X.
```

```python
plt.figure(figsize=(7,3))
plt.plot(X,Y,'o')
plt.plot(X,y_predict,'k-')
plt.show()
# plt.figure(figsize=(7,3)): Создает графическое окно (фигуру) размером 7 дюймов в ширину и 3 дюйма в высоту.
# plt.plot(X, Y, 'o'): Строит график исходных данных. 
#X и Y — это, вероятно, массивы или Series pandas, содержащие значения для осей X и Y соответственно. 
#Символ 'o' указывает, что точки данных будут отображаться в виде кружков.
# plt.plot(X, y_predict, 'k-'): Строит график линии регрессии. X — те же значения, что и на предыдущем графике, а
#y_predict — массив предсказанных значений зависимой переменной, полученных с помощью модели линейной регрессии. 
#'k-' указывает, что линия будет черного цвета (k — black) и сплошной ( - — solid line).
#plt.show(): Отображает созданный график.

```

```python
from sklearn.metrics import mean_squared_error, r2_score
#sklearn — это библиотека машинного обучения. metrics — модуль, содержащий различные метрики для оценки качества моделей.
#mean_squared_error: Импортирует функцию для вычисления среднеквадратичной ошибки (Mean Squared Error, MSE). 
#MSE — это распространённая метрика, измеряющая среднее квадратов разностей между фактическими и предсказанными значениями. 
#Чем меньше MSE, тем лучше модель соответствует данным.
#r2_score: Импортирует функцию для вычисления коэффициента детерминации (R-квадрат, R²).
#R² показывает, какую долю дисперсии зависимой переменной объясняет модель. 
#Значение R² находится в диапазоне от 0 до 1, где 1 означает идеальное соответствие модели данным.
# В итоге, после выполнения этой строки кода, вы сможете использовать функции mean_squared_error 
#r2_score для оценки качества вашей модели линейной регрессии.

```

```python
mean_squared_error(Y,y_predict)
# mean_squared_error(Y, y_predict) — это функция из библиотеки scikit-learn в Python, которая вычисляет среднеквадратичную ошибку (MSE)
#между фактическими значениями (Y) и предсказанными значениями (y_predict).
```

```python
r2_score(Y,y_predict)
# r2_score(Y, y_predict) — это функция из библиотеки scikit-learn в Python, 
# которая вычисляет коэффициент детерминации (R-квадрат) для оценки качества модели.
```
