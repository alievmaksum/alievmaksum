```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
df_pai3= pd.read_csv("https://raw.githubusercontent.com/artamonoff/econometrica/refs/heads/main/econometrica-1/datasets/Mishkin.csv")
df_pai3
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.scatter (xs = 'pai1', ys = 'pai3',zs = 'tb3', data = df_pai3)
plt.show()
```

```python
#Подгоним плоскость 

fitted_plane = smf.ols(formula= 'pai3 ~ pai1 + np.log(tb3)', data=df_pai3).fit()
```

```python
fitted_plane.params
```

```python
fitted_plane = smf.ols(formula= 'pai3 ~ pai1 + np.cos(tb3)', data=df_pai3).fit()
```

```python
fitted_plane.params
```

```python
beta0, beta1, beta2 = fitted_plane.params
```

```python
X= np.arange(start = df_pai3 ['pai1'].min(), stop = df_pai3['pai1'].max(), step=1)
Y= np.arange(start = df_pai3 ['tb3'].min(), stop = df_pai3 ['tb3'].max(), step=1)
X,Y = np.meshgrid(X,Y)
```

```python
Z = beta0 + beta1 * X+ beta2 * Y
```

```python
fig.ax = plt.subplots(subplot_kw = {'projection': '3d'})
ax.plot_surface(X,Y,Z, color = 'r', alpha = 0.2)
ax.scatter (zs = 'tb3', xs = 'pai1', ys = 'pai3', data = df_pai3)
plt.show
```
